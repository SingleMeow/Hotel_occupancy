```python
import os
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '1'

from data_prep import *
from functions import *
import parameters as params

STEPS = params.STEPS
N_FEATURES = params.N_FEATURES
```

```python
npzfile = np.load('data_numpy.npz')
X_train, y_train = npzfile['arr_0'], npzfile['arr_1']
```

```python
model = create_model(neuron_type = "LSTM", n_cells = 2, learn_rate = 0.001)
callbacks = [My_Callback(model.name)]
```

```python
history = model.fit(
    x = X_train,
    y = y_train, 
    verbose = 0,
    validation_split = 0.15,
    shuffle = True, 
    epochs = 100, 
    batch_size = 300,
    callbacks = callbacks)
```
